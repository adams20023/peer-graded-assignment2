```python
# Import necessary libraries
import yfinance as yf
import pandas as pd

# Create a Ticker object for Tesla
tsla = yf.Ticker("TSLA")

# Download historical data with the maximum available period
tesla_data = tsla.history(period="max")

# Reset the index to turn the date into a column
tesla_data.reset_index(inplace=True)

# Display the first five rows of the dataframe
print(tesla_data.head())

# Save the dataframe to a CSV file (optional)
tesla_data.to_csv("tesla_stock_data.csv", index=False)
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd

# Set up Chrome options with headless mode
chrome_options = Options()
chrome_options.binary_location = "/Applications/Google Chrome.app/Contents/MacOS/Google Chrome"
chrome_options.add_argument("--headless")  # Run Chrome in headless mode
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Use WebDriverManager to automatically download and manage the right version of ChromeDriver
service = Service(ChromeDriverManager().install())

# Initialize the driver with service and options
driver = webdriver.Chrome(service=service, options=chrome_options)

# Open Yahoo Finance page for Tesla
driver.get("https://finance.yahoo.com/quote/TSLA/financials?p=TSLA")

try:
    # Wait for the financial data section to load
    WebDriverWait(driver, 30).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "div[data-test='fin-row']"))
    )
    print("Page loaded successfully.")

    # Get the page source and parse it with BeautifulSoup
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # Find and extract the financial data
    table_rows = soup.find_all('div', {'data-test': 'fin-row'})
    revenue = []
    dates = ["2023", "2022", "2021", "2020", "2019"]

    for row in table_rows:
        spans = row.find_all('span')
        if spans:
            metric = spans[0].text.strip()
            if metric == "Total Revenue":
                for span in spans[1:]:
                    revenue.append(span.text.strip())
                break

    # Adjust lengths to match
    min_length = min(len(dates), len(revenue))
    dates = dates[:min_length]
    revenue = revenue[:min_length]

    # Create a dataframe
    tesla_revenue = pd.DataFrame({
        "Year": dates,
        "Total Revenue": revenue
    })

    # Display the dataframe
    print(tesla_revenue)

except Exception as e:
    print(f"An error occurred: {e}")

finally:
    # Close the browser
    driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import pandas as pd

# Set up Chrome options (optional headless mode)
chrome_options = Options()
chrome_options.add_argument("--headless")  # Run in headless mode if you don't want the browser to open
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")

# Initialize ChromeDriver using WebDriverManager
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=chrome_options)

# URL of GameStop's financials page on Yahoo Finance
url = "https://finance.yahoo.com/quote/GME/financials?p=GME"

# Open the URL in the browser using Selenium
driver.get(url)

# Wait for the page to load and get the page source
page_source = driver.page_source

# Close the driver after getting the page source
driver.quit()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(page_source, 'html.parser')

# Find all table rows in the financials section
table_rows = soup.find_all('div', {'data-test': 'fin-row'})

# Initialize lists to store data
revenue = []
dates = ["2023", "2022", "2021", "2020", "2019"]  # Adjust based on available data

for row in table_rows:
    # Find all spans within the row
    spans = row.find_all('span')
    if spans:
        # The first span is the metric name
        metric = spans[0].text.strip()
        if metric == "Total Revenue":
            # Extract revenue figures (skip the first span as it is the metric name)
            for span in spans[1:]:
                revenue.append(span.text.strip())
            break  # Exit after finding Total Revenue

# Check if revenue data was found
if len(revenue) == 0:
    print("No revenue data found.")
else:
    # Adjust the length of dates and revenue to match
    min_length = min(len(dates), len(revenue))
    dates = dates[:min_length]
    revenue = revenue[:min_length]

    # Create a dataframe
    gme_revenue = pd.DataFrame({
        "Year": dates,
        "Total Revenue": revenue
    })

    # Display the last five rows of the dataframe
    print(gme_revenue.tail())

    # Save the dataframe to a CSV file (optional)
    gme_revenue.to_csv("gme_revenue.csv", index=False)
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Load Tesla stock data
tesla_data = pd.read_csv("tesla_stock_data.csv")

# Load Tesla revenue data
tesla_revenue = pd.read_csv("tesla_revenue.csv")

# Convert 'Date' to datetime, handling errors and setting UTC to handle timezone issues
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'], errors='coerce', utc=True)

# Check for any rows with missing dates and drop them if necessary
tesla_data = tesla_data.dropna(subset=['Date'])

# Extract year from 'Date'
tesla_data['Year'] = tesla_data['Date'].dt.year

# Group stock data by year and calculate average closing price
annual_stock = tesla_data.groupby('Year')['Close'].mean().reset_index()

# Merge with revenue data
dashboard_data = pd.merge(tesla_revenue, annual_stock, on='Year')

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Total Revenue on primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Revenue (in millions USD)', color='tab:blue')
ax1.plot(dashboard_data['Year'], dashboard_data['Total Revenue'], color='tab:blue', marker='o', label='Total Revenue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a secondary y-axis for Stock Price
ax2 = ax1.twinx()
ax2.set_ylabel('Average Closing Stock Price (USD)', color='tab:red')
ax2.plot(dashboard_data['Year'], dashboard_data['Close'], color='tab:red', marker='x', label='Avg Close Price')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Add a title
plt.title('Tesla Stock Price vs Revenue')

# Add legends
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))

# Show plot
plt.show()
```

```python
# Import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd

# Use the correct file paths for your data
tesla_stock_data_path = "tesla_stock_data.csv"  # Correct path for Tesla stock data
tesla_revenue_data_path = "tesla_revenue.csv"   # Correct path for Tesla revenue data

# Load Tesla stock data
tesla_data = pd.read_csv(tesla_stock_data_path)

# Load Tesla revenue data
tesla_revenue = pd.read_csv(tesla_revenue_data_path)

# Convert 'Date' to datetime, handling errors and specifying utc=True to handle mixed timezones
tesla_data['Date'] = pd.to_datetime(tesla_data['Date'], errors='coerce', utc=True)

# Drop rows with invalid dates (if any)
tesla_data = tesla_data.dropna(subset=['Date'])

# Extract year from 'Date'
tesla_data['Year'] = tesla_data['Date'].dt.year

# Group stock data by year and calculate average closing price
annual_stock_tesla = tesla_data.groupby('Year')['Close'].mean().reset_index()

# Ensure both datasets have matching years
dashboard_data_tesla = pd.merge(tesla_revenue, annual_stock_tesla, on='Year', how='inner')

# Create a figure and axis
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Total Revenue on primary y-axis
ax1.set_xlabel('Year')
ax1.set_ylabel('Total Revenue (in millions USD)', color='tab:green')
ax1.plot(dashboard_data_tesla['Year'], dashboard_data_tesla['Total Revenue'], color='tab:green', marker='o', label='Total Revenue')
ax1.tick_params(axis='y', labelcolor='tab:green')

# Create a secondary y-axis for Stock Price
ax2 = ax1.twinx()
ax2.set_ylabel('Average Closing Stock Price (USD)', color='tab:orange')
ax2.plot(dashboard_data_tesla['Year'], dashboard_data_tesla['Close'], color='tab:orange', marker='x', label='Avg Close Price')
ax2.tick_params(axis='y', labelcolor='tab:orange')

# Add a title
plt.title('Tesla Stock Price vs Revenue')

# Add legends for each axis
ax1.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9))
ax2.legend(loc='upper right', bbox_to_anchor=(0.9, 0.9))

# Show plot
plt.show()
```
